### Exercici 1
#### Crea un np.array d'una dimensió, que inclogui l'almenys 8 nombres sencers, data type int64. Mostra la dimensió i la forma de la matriu.

```python
import numpy as np #importa el paquete numpy
matriu = np.array([15,33,45,21,89,62,90,76], dtype = np.int64) #creación de matriz np
print("la matriz es:",matriu)
print ("la dimensión de la matriz es:", matriu.ndim) #imprime las dimensiones de la matriz
print ("la forma de la matriz es:", matriu.shape) #imprime la forma de la matriz
```

### Exercici 2
#### De la matriu de l'exercici 1, calcula el valor mitjà dels valors introduïts i resta la mitjana resultant de cada un dels valors de la matriu.

```python
valor_mitja = matriu.mean() #método mean calcula el valor medio
print ("el valor medio de la matriz es:", valor_mitja,"\n") 
matriu = matriu - valor_mitja
print ("resta de la media a los valores de la matriz:\n", matriu)
```

### Exercici 3
#### Crea una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

```python
matriu_bid = np.random.randint(1,100,(5,5)) #matriz de números aleatorios 1-100
print ("matriz de 5x5 generada aleatoriamente:\n", matriu_bid, "\n") #imprime la matriz

print ("el valor máximo de la matriz es", matriu_bid.max(),"\n") #imprime el máximo valor de la matriz

i = 0
while i < 5:
    print ("el valor máximo por columna eje" ,i+1, "es:", matriu_bid[:,i].max()) #imprime el valor máximo de cada columna
    print ("el valor máximo por fila    eje" ,i+1, "es:", matriu_bid[i,:].max()) #imprime el valor máximo de cada columna
    i=i+1
```

### Exercici 4
#### Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
matriu_1 = np.array([[8,2,3],[3,7,9]])
print("la matriz A es:\n", matriu_1)
print ("la dimensión de la matriz A es", matriu_1.ndim,"y tamaño", matriu_1.shape, "\n")
matriu_2 = np.array([[4,5,9],[5,6,1]])
print("la matriz B es:\n", matriu_2)
print ("la dimensión de la matriz B es", matriu_2.ndim,"y tamaño", matriu_2.shape, "\n")
matriu_3 = matriu_1 * matriu_2 #compatibles porque tienen la misma dimensión
print ("el resultado de multiplicar dos matrices con igual tamaño en la dimensión es:\n", matriu_3,"\n")

matriu_4 = np.array([[4,6,8],[2,9,1],[3,6,9]])
print("la matriz X es:\n", matriu_4)
print ("la dimensión de la matriz X es", matriu_4.ndim,"y tamaño", matriu_4.shape, "\n")
matriu_5 = np.array([10])
print("la matriz Y es:\n", matriu_5)
print ("la dimensión de la matriz Y es", matriu_5.ndim,"y tamaño", matriu_5.shape, "\n")
matriu_6 = matriu_4 * matriu_5 #compatibles porque porque una matriz es de tamaño 1
print ("el resultado de multiplicar con una matriz de tamaño 1 en una dimensión es:\n", matriu_6)
```

### Exercici 8

#### Manipulació d’imatges amb Matplotlib.

#### Carregareu qualsevol imatge (jpg, png ..) amb Matplotlib. adoneu-vos que les imatges RGB (Red, Green, Blue) són realment només amplades × alçades × 3 matrius (tres canals Vermell, Verd i Blau), una per cada color de nombres enters int8, manipuleu aquests bytes i torneu a utilitzar Matplotlib per desar la imatge modificada un cop hàgiu acabat.

#### Ajuda:Importeu, import matplotlib.image as mpimg. estudieu el metodde mpimg.imread(()

```python
import matplotlib.image as mpimg #importa el módulo de manipulación de imágenes
import matplotlib.pyplot as plt #importa el módulo de dibujo de imágenes

homer = mpimg.imread('homer.png') #importa la imagen y la convierte en matriz
print ("la matriz tiene el tamaño (alturaxanchoxRGB):", homer.shape) #imprime el tamaño de la matriz
print ("la dimensión de la matriz de la imagen es:", homer.ndim) #imprime dimensiones de la matriz
imgplot = plt.imshow(homer) #imprime la imagen a partir de la matriz
```

#### Mostreu-me a veure què passa quan eliminem el canal G Verd o B Blau. Hauries d'utilitzar la indexació per seleccionar el canal que voleu anul·lar.


```python
homer_noG = mpimg.imread('homer.png') #importa la imagen y la convierte en matriz
homer_noG[:,:,1] = 0 #elimina componente G (verde)
print ("imagen sin componente G (verde):")
imgplot = plt.imshow(homer_noG) #imprime la imagen
```

```python
homer_noB = mpimg.imread('homer.png') #importa la imagen y la convierte en matriz
homer_noB[:,:,2] = 0 #elimina componente B (azul)
print ("imagen sin componente B (azul):")
imgplot = plt.imshow(homer_noB) #imprime la imagen
```

#### Utilitzar el mètode, mpimg.imsave () de la llibreria importada, per guardar les imatges modificades i que haureu de pujar al vostre repositori a github.

```python
plt.imsave("homer_noG.png", homer_noG) #guarda la imagen en la raiz del cuaderno con nombre parámetro1 de la imagen parámetro2
plt.imsave("homer_noB.png", homer_noB) 
```
